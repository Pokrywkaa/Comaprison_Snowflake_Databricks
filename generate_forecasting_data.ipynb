```python
import pandas as pd
import numpy as np


rows_number = 1000
# Seed for reproducibility
np.random.seed(42)

# Generate dates
dates = pd.date_range(start='2020-01-01', periods=rows_number, freq='D')

# Generate data based on a sine function with noise and a linear trend
time = np.arange(rows_number)
sine_wave = 20 * np.sin(2 * np.pi * time / 365)  # yearly cycle
noise = np.random.normal(scale=5, size=rows_number)  # random noise
trend = time * 0.05  # simple linear trend

values = sine_wave + noise + trend

# outliers = np.random.choice([False, True], size=rows_number, p=[0.99, 0.01])
# values[outliers] = values[outliers] * np.random.uniform(1.5, 3.0)

# Create DataFrame
df = pd.DataFrame({
    'date': dates,
    'value': values
})

# Print the first few rows of the DataFrame
print(df.head())
```

```python
df.to_csv('generated_data/forecasting_1k.csv', index=False)
```

```python
# Plotting
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6))
plt.plot(df['date'], df['value'], label='value', color='blue')
plt.title('Generated Test Data: 1000')
plt.xlabel('date')
plt.ylabel('value')
plt.legend()
plt.grid(True)
plt.show()
```
